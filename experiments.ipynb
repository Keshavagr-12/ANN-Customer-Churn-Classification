```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
df = pd.read_csv('Churn_Modelling.csv')
df
```

```python
df = df.drop(['RowNumber','CustomerId','Surname'],axis=1)
df
```

```python
label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])
df
```

```python
df['Geography'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder = OneHotEncoder()

geo_encoder = one_hot_encoder.fit_transform(df[['Geography']]).toarray()
one_hot_encoder.get_feature_names_out(['Geography'])
```

```python
geo_encoded_df=pd.DataFrame(geo_encoder,columns=one_hot_encoder.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
df =pd.concat([df.drop('Geography',axis=1),geo_encoded_df],axis=1)
```

```python
with open('label_encoder.pkl','wb') as file:
    pickle.dump(label_encoder, file)

with open('one_hot_encoder.pkl','wb') as file:
    pickle.dump(one_hot_encoder, file)
```

```python
X = df.drop('Exited',axis=1)
y = df['Exited']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
model = Sequential([
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)),
    Dense(32,activation='relu'),
    Dense(1,activation='sigmoid')
])
```

```python
model.summary()
```

```python
import tensorflow
opt = tensorflow.keras.optimizers.Adam(learning_rate=0.01)
```

```python
model.compile(optimizer="Adam",loss="binary_crossentropy",metrics=['accuracy'])
```

```python
log_dir = "logs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorflow_callbacks = TensorBoard(log_dir = log_dir,histogram_freq=1)
```

```python
EarlyStopping_callbacks = EarlyStopping(monitor='val_loss',patience=10, restore_best_weights=True)
```

```python
history = model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs = 100, callbacks=[tensorflow_callbacks,EarlyStopping_callbacks])
```

```python
import pickle

with open("model.pkl", "wb") as f:
    pickle.dump(model, f)
```
